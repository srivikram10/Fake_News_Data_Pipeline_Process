```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    r'C:\My Projects\fake_news_pipeline\data\processed\clean_news.csv'
)
```

```python
print("Shape:", df.shape)

df.head()

df.info()

df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
null_counts = df.isnull().sum()

plt.figure(figsize=(8,4))
plt.bar(null_counts.index, null_counts.values, label="Missing Values")
plt.title("Missing Values Analysis")
plt.xlabel("Columns")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
counts = df['label'].value_counts()

plt.figure(figsize=(6,4))

plt.bar(
    ['Fake News', 'Real News'],
    counts.values,
    label='News Articles'
)

plt.title("Fake vs Real News Distribution")
plt.xlabel("News Type")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
top_subjects = df['subject'].value_counts().head(10)

plt.figure(figsize=(10,5))

plt.bar(
    top_subjects.index,
    top_subjects.values,
    label="Articles"
)

plt.title("Top 10 News Categories")
plt.xlabel("Subject")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.legend()
plt.show()
```

```python
df['text_length'] = df['text'].str.len()
```

```python
df['text_length'].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df['text_length'],
    bins=30,
    label="Article Length"
)

plt.title("News Length Distribution")
plt.xlabel("Characters")
plt.ylabel("Frequency")

plt.legend()
plt.show()
```

```python
avg_length = df.groupby('label')['text_length'].mean()

avg_length
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    ['Fake News', 'Real News'],
    avg_length.values,
    label='Average Length'
)

plt.title("Average News Length")
plt.xlabel("News Type")
plt.ylabel("Average Characters")

plt.legend()
plt.show()
```
